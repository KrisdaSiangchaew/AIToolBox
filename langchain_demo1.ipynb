```python
from langchain.agents import Tool
from langchain_community.utilities import GoogleSerperAPIWrapper

from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
```

```python
serper = GoogleSerperAPIWrapper()
```

```python
# directly via serper
serper.run('what is capital of mongolia')
```

```python
# via langchain way
tool_search = Tool(name='search', func=serper.run, description='userful for online search')
tool_search.invoke('What is capital of sri lanka')
```
